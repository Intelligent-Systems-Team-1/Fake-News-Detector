```python
from urlextract import URLExtract

extractor = URLExtract()

def remove_urls(text):
    urls = extractor.find_urls(text)
    for url in urls:
        text = text.replace(url, '')
    return text.strip()

def clean_text(text, bert=False):
    if not bert:
        text = str(text).lower()
    text = remove_urls(text)
    if not bert:
        text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()

def clean_true_and_fake_data(true_df, fake_df, bert=False):
    true_df = true_df.copy()
    true_df.dropna(subset=["title", "text"], inplace=True)
    true_df["text_combined"] = true_df["title"] + " " + true_df["text"]
    true_df["label"] = 1
    true_df = true_df[["text_combined", "label"]]

    fake_df = fake_df.copy()
    fake_df.dropna(subset=["title", "text"], inplace=True)
    fake_df["text_combined"] = fake_df["title"] + " " + fake_df["title"]
    fake_df["label"] = 0
    fake_df = fake_df[["text_combined", "label"]]

    combined_df = pd.concat([true_df, fake_df], ignore_index=True)
    combined_df["text_combined"] = combined_df["text_combined"].apply(lambda s: clean_text(s, bert=bert))
    combined_df["text_combined"] = combined_df["text_combined"].replace('', pd.NA)
    combined_df.rename(columns={"text_combined": "text"}, inplace=True)
    combined_df.dropna(inplace=True)

    return combined_df

def clean_liar_dataset(liar_df, binary=True, bert=False):
    liar_df = liar_df.copy()
    liar_df.dropna(subset=["statement", "label"], inplace=True)

    label_map_binary = {
        'true': 1,
        'mostly-true': 1,
        'half-true': 1,
        'barely-true': 0,
        'false': 0,
        'pants-fire': 0
    }

    label_map = {
        'true': 1,
        'mostly-true': 2,
        'half-true': 3,
        'barely-true': 4,
        'pants-fire': 5,
        'false': 0
    }

    if binary:
        liar_df['label'] = liar_df['label'].str.lower().map(label_map_binary)
    else:
        liar_df['label'] = liar_df['label'].str.lower().map(label_map)
    liar_df.dropna(subset=['label'], inplace=True)
    liar_df['label'] = liar_df['label'].astype(int)

    liar_df['text_cleaned'] = liar_df['statement'].apply(lambda s: clean_text(s, bert=bert))
    liar_df['text_cleaned'] = liar_df['text_cleaned'].replace('', pd.NA)
    liar_df = liar_df[['text_cleaned', 'label']]
    liar_df.rename(columns={"text_cleaned": "text"}, inplace=True)
    liar_df.dropna(inplace=True)

    return liar_df
```

```python
df1 = clean_true_and_fake_data(true_df=true_df, fake_df=fake_df)
df2 = clean_liar_dataset(liar_df=liar_df)
combined_binary_df = pd.concat([df1, df2], ignore_index=True)
combined_binary_df.to_csv("../cleaned_datasets/cleaned_news_binary.csv", index=False)

df3 = clean_true_and_fake_data(true_df=true_df, fake_df=fake_df, bert=True)
df4 = clean_liar_dataset(liar_df=liar_df, bert=True)
combined_binary_df_bert = pd.concat([df3, df4], ignore_index=True)
combined_binary_df_bert.to_csv("../cleaned_datasets/cleaned_news_binary_bert.csv", index=False)

df6 = clean_liar_dataset(liar_df, binary=False)
combined_multinomial_df = pd.concat([df1, df6], ignore_index=True)
combined_multinomial_df.to_csv("../cleaned_datasets/cleaned_news_multinomial.csv", index=False)

df7 = clean_liar_dataset(liar_df=liar_df, binary=False, bert=True)
combined_multinomial_df_bert = pd.concat([df3, df7], ignore_index=True)
combined_multinomial_df_bert.to_csv("../cleaned_datasets/cleaned_news_multinomial_bert.csv", index=False)
```
